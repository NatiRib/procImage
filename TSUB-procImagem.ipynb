```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

# Exercicio 1

```python
img = cv2.imread('dog.png', 0)
plt.imshow(img, cmap = 'gray')
plt.title('Imagem original'), plt.xticks([]), plt.yticks([])
plt.show()
```

Função que recebe uma imagem em escala cinza e gera uma imagem binarizada

```python
ret, gray = cv2.threshold(img, 127, 256, cv2.THRESH_BINARY)
```

```python
plt.imshow(gray, cmap = 'gray')
plt.title('Imagem binarizada')
plt.show()
```

# Exercicio 2

```python
im2, contours, hierarchy = cv2.findContours(gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img, contours, -1, (255,0,0), 3)
plt.title('Bordas de imagem binarizadas')
plt.imshow(img)
```
