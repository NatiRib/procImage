```python
def surface_plot (matrix, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    (x, y) = np.meshgrid(np.arange(matrix.shape[0]), np.arange(matrix.shape[1]))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x, y, matrix, **kwargs)
    return (fig, ax, surf)
```

# Exercício 3

```python
def circulo(modo, altura, largura, centro=None, raio=None):

    if centro is None: # Centro da imagem
        centro = [int(largura/2), int(altura/2)]
    if raio is None: # Distancia entre o centro e as bordas da imagem
        raio = min(centro[0], centro[1], largura-centro[0], altura-centro[1])

    Y, X = np.ogrid[:largura, :altura]
    dist_from_centro = np.sqrt((X - centro[0])**2 + (Y-centro[1])**2)
    
    if modo == 0: #Filtro Passa Baixa
        mask = dist_from_centro <= raio
        return mask
    if modo == 1: #Filtro Passa Alta
        mask = dist_from_centro >= raio
        return mask
```

```python
imagem = cv2.imread("CARRO.BMP",0)

plt.title("Original")
plt.imshow(imagem)
plt.show()

#Gera uma nova imagem com dimensões 128x128
imagem2 =  imagem[200:328, 200:328]
altura2, largura2  = imagem2.shape

plt.title("Placa enquadrada")
plt.imshow(imagem2)
plt.show()

temp_imagem = np.zeros((256,256))
    
#Ampliacao sem filtro (Tem que varrer a imagem para intercalar cada pixel)

y=0
x=0
    
for i in range(128):
    for j in range(128):
             temp_imagem[(x,y)] = imagem2[(j,i)]
             x+=2
    x=0
    y+=2

plt.title("Ampliacao Sem Filtro")
plt.imshow(temp_imagem)
plt.show()
    
#Realiza a transformada de fourier bidimensional, shift e aplica a magnitute
f = np.fft.fft2(temp_imagem) 
fshift = np.fft.fftshift(f) 
magnitude_spectrum = np.log(1+ 0.01*np.abs(fshift)) 
   
#Filtro Passa Baixa Ideal
altura, largura  = temp_imagem.shape
center = [int(altura/2), int(largura/2)]
m = circulo(0,altura,largura,center,70)
    
m = m.transpose()
    
temp = np.multiply(magnitude_spectrum,m)
figure(num=None, figsize=(9, 9), dpi=80, facecolor='w', edgecolor='k')

plt.title("Filtro Passa Baixa Ideal")
plt.imshow(temp)
plt.show()
    
#Convolucao do Filtro Passa Baixa Ideal com o Shift amplificado
t = np.multiply(m,fshift)
    
t = np.fft.ifftshift(t)

t = np.fft.ifft2(t)
   
t = np.abs(t)
    
plt.title("Ampliacao com Filtro")
plt.imshow(t)
plt.show()
```
